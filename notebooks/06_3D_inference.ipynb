Estimate the relative 3D position of a section within the pancreas based on simple heuristics.

```
import cv2
from isletscope.spatial_inference import SpatialInferer

image_path = 'path/to/your/image.tif'
img = cv2.imread(image_path)
if img is None:
    raise FileNotFoundError(f'Cannot load image at {image_path}')

inferer = SpatialInferer()
pos = inferer.infer(img)
print('Estimated normalized coordinates:')
for key, val in pos.items():
    print(f'{key}: {val:.3f}')
```
